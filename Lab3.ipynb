## Example code

```python
#Define Parameters
w = torch.zeros(1, requires_grad = True)    # Always True!
b = torch.zeros(1, requires_grad = True)

# Define dataset using tensor
x_train = torch.FloatTensor([[78],[83],[56],[67],[85],[44],[32],[90]])
y_train = torch.FloatTensor([[66],[73],[76],[65],[81],[54],[29],[85]])

# Define optimizer using SGD
optimizer = optim.SGD([w,b], lr = 0.0001)

# Optimization
num_epochs = 1000
costs = []

for epoch in range(num_epochs):
    hypothesis = x_train * w + b

    # Define cost function using MSE
    cost = torch.mean((hypothesis - y_train) ** 2)
    optimizer.zero_grad()
    cost.backward()
    optimizer.step()

    print("Epoch: %d  cost: %f" %(epoch,cost))

# evaluation
print("\nw: %f, b: %f" %(w,b))
test_value = 71
predict = w * test_value + b
print('\nmy final score is estimated sd %d' %(predict))
```

```python
import numpy as np
import pandas as pd
import torch
import torch.nn as nn
import torch.optim as optim
import torch.nn.init as init
# Define Parameters
w1 = torch.zeros(1, requires_grad = True)    # Always True!
w2 = torch.zeros(1, requires_grad = True)    # Always True!
w3 = torch.zeros(1, requires_grad = True)    # Always True!
w4 = torch.zeros(1, requires_grad = True)    # Always True!
b = torch.zeros(1, requires_grad = True)

# Define dataset using tensor
x_GPA = torch.FloatTensor([[3.8],[3.2],[3.7],[4.2],[2.6],[3.4],[4.1],[3.3],[3.9]])
x_TOEIC = torch.FloatTensor([[700],[650],[820],[830],[550],[910],[990],[870],[650]])
x_Award = torch.FloatTensor([[80],[90],[70],[50],[90],[30],[70],[60],[80]])
x_Etc = torch.FloatTensor([[50],[30],[40],[70],[60],[40],[20],[60],[50]])

test_GPA = 3.3
test_TOEIC = 700
test_Award = 77
test_Etc = 84

y_train = torch.FloatTensor([[85],[80],[78],[87],[85],[70],[81],[88],[84]])

# Define optimizer using SGD
optimizer = optim.SGD([w1,w2,w3,w4,b], lr = 0.000001)

# Optimization
num_epochs = 3000

for epoch in range(num_epochs):
    hypothesis = x_GPA * w1 + x_TOEIC * w2 + x_Award * w3 + x_Etc * w4 + b

    # Define cost function using MSE
    cost = torch.mean((hypothesis - y_train) ** 2)
    optimizer.zero_grad()
    cost.backward()
    optimizer.step()
    costs.append(cost.item())

    print("Epoch: %d  cost: %f" %(epoch,cost))
    #print('%f %f %f %f %f' %(w1,w2,w3,w4,b))

# evaluation
predict = test_GPA * w1 + test_TOEIC * w2 + test_Award * w3 + test_Etc * w4 + b
print('\nmy final score is estimated sd %d' % (predict))

# Visualization of output
plt.plot(range(num_epochs), costs)
plt.title('Training Curve')
plt.xlabel('Epoch')
plt.ylabel('Cost')
plt.show()
```
